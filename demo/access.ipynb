## data access demo
best to run from VSCode and use the same virtual environment you use for coding (`.venv` if you use `uv` )

```python
import dlt_portable_data_lake_demo
```

```python
catalog_ = dlt_portable_data_lake_demo.catalog()
raw_data_ = catalog_.github_events_dataset
print(raw_data_)
```

```python
# get shape
issues_ = raw_data_.issues_event
issues_.head().df()
```

```python
# use ibis to generate SQL statement executed via duckdb on a view of Iceberg table 🤯
# below we filter out all github action bot issues
closed_issues = issues_.filter(issues_.actor__display_login != "github-actions")
closed_issues.df()
```

```python
# get duckdb connection and access iceberg views directly
import duckdb

with raw_data_.sql_client as c:
    conn: duckdb.DuckDBPyConnection = c.native_connection
    print(conn.sql("SHOW TABLES;"))
    print(conn.sql("SELECT * FROM issues_event LIMIT 5"))
```

```python
# get direct pyiceberg access to the datalake
catalog_ = dlt_portable_data_lake_demo.catalog(profile="iceberg_modernization")
raw_data_ = catalog_.github_events_dataset

iceberg_table = raw_data_.table_client.load_open_table("iceberg", "issues_event")
print(iceberg_table.location())
print(iceberg_table.scan().to_arrow())
```

```python
# show catalog info
print("CATALOG INFO")
iceberg_catalog = raw_data_.table_client.get_open_table_catalog("iceberg")
print(iceberg_catalog.list_namespaces())
```
